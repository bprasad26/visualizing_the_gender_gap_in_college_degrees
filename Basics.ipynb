```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
sns.set_style('white')
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_degrees.head(10)
```

# Visualize the Data 

```python
# Plot a line chart a line chart that visualizes the 
# historical percentage of Biology degrees awarded to women.
plt.plot(women_degrees['Year'], women_degrees['Biology']);
```

### Gender Gap in Biology Degrees

```python
plt.plot(women_degrees['Year'], women_degrees['Biology'], c='blue', label='Women')
plt.plot(women_degrees['Year'], 100-women_degrees['Biology'], c='green', label='Men')
plt.legend(loc='upper right')
plt.title('Percentage of Biology Degrees Awarded By Gender')
plt.show()
```

Looking at the above figure, we can see that in the first period, from 1970 to around 1987, women were a minority 
when it came to majoring in Biology while in the second period, from around 1987 to around 2012, women became 
a majority. You can see the point where women overtook men where the lines intersect.

```python
stem_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics','Psychology','Physical Sciences']
fig = plt.figure(figsize=(12, 12))

for sp in range(0,6):
    ax = fig.add_subplot(3,2,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c='blue', label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c='green', label='Men')

    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    ax.set_title(stem_cats[sp])
    
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

# Calling pyplot.legend() here will add the legend to the last subplot that was created.
plt.legend(loc='upper right')
plt.show()
```

**Few Things we can observe from the above figures:**
* Biology - Initially the percentage of Men in Biology were higher than Female but after near around 1986-87, the trend started to get reversed. The men percentages start to keep getting lower and female percentages started to 
    get higher higher.
* There is always a higher gender gap in Engineering and Computer Science, though some gap is reduced in computer 
    science than again is started to widen. Men always tends to dominate in these two fields.
* There is also a gender gap in Math and statistics but is not as wide as compared to Engineering and Computer 
    science.
* When it comes to Psychology,the percentage of women is always way higher than Men. And In Physical sciences it again opposite, though gap narrowed down later.
    

## Comparing across all degree categories

```python
stem_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics','Psychology','Physical Sciences']

lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 
                 'Social Sciences and History']

other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
# For Color blindness
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

# Create the size of the plot
fig = plt.figure(figsize=(16, 20))

# Generate first column of line charts. STEM degrees.
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if cat_index == 0:
        ax.text(2003, 65, 'Women')
        ax.text(2005, 25, 'Men')
    elif cat_index == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2003, 30, 'Women')
        
# Generate second column of line charts. Liberal arts degrees.
for sp in range(1,16,3):
    cat_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if cat_index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
    
## Generate third column of line charts. Other degrees.
for sp in range(2,20,3):
    cat_index = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if cat_index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif cat_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        
plt.show()
```
